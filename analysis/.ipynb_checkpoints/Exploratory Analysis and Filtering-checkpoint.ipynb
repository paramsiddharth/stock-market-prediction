```python
# Load modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read data from CSV
data = pd.read_csv('reliance.csv', thousands=',', dtype={
    'Date': 'str',
    'Price': 'float',
    'Open': 'float',
    'High': 'float',
    'Low': 'float',
    'Vol': 'float',
    'Chg': 'float',
    'crude oil price': 'float',
    'dollar': 'float'
})
```

```python
# First look at data
data.head()
```

```python
# Datatypes
data.dtypes
```

```python
# NA counts
pd.isna(data[['Date', 'Price', 'Open', 'High', 'Low', 'Volume', 'Vol', 'Chg', 'crude oil price', 'dollar']]).sum()
```

```python
# Parsing volume from 'Volume'
def parse_volume(volume):
    M_index = volume.index('M')
    volume = volume[:M_index]
    return pd.to_numeric(volume) * 1_000_000

data['Volume'] = data['Volume'].map(parse_volume)

# Turns out it has already been parsed in 'Vol' in the unit of millions, I'll drop this one
```

```python
data.dtypes
```

```python
# Testing datetime conversion of 'Date'
pd.to_datetime(data['Date'][0])
```

```python
#pd.to_datetime(data['Date']) # Fails due to dirty data
```

```python
# Some of the data is polluted with unwanted characters
# Extract a fixed length of 12 characters from each 'Date' string
data['Date'] = data['Date'].map(lambda s: s[:12])
```

```python
# Parse dates successfully
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.dtypes
```

```python
data.drop('Volume', axis=1, inplace=True)
```

```python
data.tail()
```

```python
data.columns = ['Date', 'Price', 'Open', 'High', 'Low', 'Volume', 'Change', 'Crude Oil Price', 'Dollar']
```

```python
data['Volume'] *= 1_000_000
data['Volume'] = data['Volume'].map(int)
```
